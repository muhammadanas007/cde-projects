```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv', encoding='utf-8')
```

```python
import sqlalchemy as sal
engine = sal.create_engine('mssql://DESKTOP-20R1LP9/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
from sqlalchemy import NVARCHAR

dtype_map = {
    'type': NVARCHAR(10),
    'title': NVARCHAR(200),
    'director': NVARCHAR(250),
    'cast': NVARCHAR(1000),
    'country': NVARCHAR(150),
    'date_added': NVARCHAR(20),
    'rating': NVARCHAR(10),
    'duration': NVARCHAR(10),
    'listed_in': NVARCHAR(100),
    'description': NVARCHAR(500)
}

df.to_sql(
    'netflix_raw',
    engine,
    if_exists='append',
    index=False,
    dtype=dtype_map,
    chunksize=500  
)
```
